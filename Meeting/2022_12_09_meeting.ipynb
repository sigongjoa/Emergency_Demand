*  현재까지의 시각화의 90%의 발생확률 일때의 기준으로 사용함  
* 현재는 약간의 데이터의 변화만 나타나더라도 발생확률을 높게 보기 때문에 살짝의 데이터의 값이 나타난다면 TFT는 이를 사건이 발생했다고 예측함
* 50%의 확률을 비교 했을 때는 ewma 패턴을 어느정도 잘 재현하는 것을 확인할 수 있음  
* quaile을 25% 혹은 50%만 사용해서  rounding을 적용하면 이전보다는 좋은 성능을 보일거 같음

### 한계점
* ewma적용하면 일종의 사건 발생 이후에 tail과 유사한 패턴을 만들어냄
* 실제 TFT의 예측은 원래의 사건 발생 시간보다 늦게 나타남
* graph 기반으로 사건의 전조 증상을 넣어서 학습 시키면 성능이 좋아질 것 같음

## meeting result

* TFT의 예측은 사건의 발생 이후에 이를 따라가는 형식임 이를 해결하기 위해서 학습을 할때 noise를 아래에 추가해서 masking을 넣는 방식도 있음


* 현재 TFT가 ewma의 패턴 자체를 학습하고 있음 이는 발생 빈도를 예측해야 하는 상황에서 적합하지 않음   
* 현재의 validation을 할 때에는 val_dataloader를 통채로 넣고 뒤에 하루치 데이터를 append하는 식으로 진행
