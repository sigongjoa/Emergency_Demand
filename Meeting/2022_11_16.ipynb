### new data
2021년도(train) -> 2022년 2월(val) 으로 1년치 데이터만 사용중  
2010년도 ~ 2020년도(train) -> 2021년도(val)으로 학습 후 결과 확인   

10년치 데이터를 사용하는 경우에는 윤년이 걸려 있어서 기존의 time_index를 추가하는 것에 문제가 있음 
* 기존의 1월 1일 -> 1, 12월 31이을 365로 바꾼후 여기에다가 년도를 곱하는 방식으로 사용  
* 시간대를 unique로 시간대만 뽑아낸 이후에 여기엗가 enumerate를 이용해서 time_idx를 만든 후 이를 같은 시간대의 여러 dong별로 time_idx를 저장  

2010-01-01 00:00:00 ~ 2021-12-31 00:00:00을 기준으로
2021-12-31 00:00:00의 time_idx= 96432이고, 이는 365(1년) * 24(시간대) * 11(11년치 데이터) + 72(윤년2일 + 12월 31일 까지 계산) 로 윤년이 포함된 time_idx임을 확인할 수 있다.  

현재 데이터를 구하는것에 어려움이 있으므로 우선 count(발생빈도) data에 대해서만 학습을 진행  

max_prediction_length, max_encoder_length는 기존과 동일한 방식으로 학습을 하고 prediction을 할 때에는 1년치(2021년)로 prediction을 확인함
