# spatial data

시간적으로 ewma를 적용해서 응급상활 발생건수(target)을 늘려보는 작업을 해봄    
하지만 그래도 target이 sparse함  

Graph를 이용해서 인접한 행정구역의 traget값을 합침
 
 1. uncolored grahp를dict로 인접합 행정구역을 정리
 2. 이를 이용해서 networkX를이용해서 그래프로 만듬
 3. networkX의 neighorbood를 이용해서 k번째 이웃한 노드들 찾음
 4. 하나의 timd index와 하나의 행정 구역에 대해서 이웃한 행정구역의 target 값을 sum함
