## 행정구역 별 포함하는 법정동  

```python
for h_dong in df['행정동\n(행정기관명)'].unique():
    print(f'행정동  : {h_dong}')
    dong_df = df[df['행정동\n(행정기관명)'] == h_dong]
    STDC_CD = dong_df['법정동'].unique()
    print(f'법정동 : {STDC_CD}')
```

## 기타 코드
shp 파일 처리  
읍/면/동 파일과 리 파일의 법정동 코드가 서로 다름  
이를 처리  

```python
df['법정동'].unique()
```

```python
# shp 파일에서 행정구역 코드만 뽑아내기
df['행정구역분류\n(7자리)'] = df['행정구역분류\n(7자리)'].astype(str)
df['행정구역분류\n(7자리)'] = df['행정구역분류\n(7자리)'].apply(lambda x: x.split('.')[0])
df['행정구역분류\n(7자리)'].unique()
```

```python
df['법정동코드'] = df['법정동코드'].astype(str)
df['법정동코드_del'] = df['법정동코드'].apply(lambda x : x[:-2])
```

```python
df[['법정동코드' , '법정동코드_del']].to_csv('ch_STDG_CD.csv' , index=False)
```
