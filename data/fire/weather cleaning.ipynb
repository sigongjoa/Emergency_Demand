```python
import os
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
pd.reset_option('all')
pd.set_option('mode.chained_assignment',  None)
import numpy as np
from scipy.stats import mode
```

```python
def get_weather_filelist(path = './weather'):
    fs = sorted(os.listdir(path))
    file_df = pd.DataFrame(columns = ['h_dong', 'file_list' , 'start' , 'end', 'file_name'])

    for f_num , f_name in enumerate(fs):
        if 'ipy' in f_name:
            pass
        else:
            f_name = f_name.split('.')[0]
            f_split = f_name.split('_')
            file_df.loc[f_num, 'h_dong'] = f_split[0]
            file_df.loc[f_num, 'file_list'] = f_split[1]
            file_df.loc[f_num, 'start'] = f_split[2]
            file_df.loc[f_num, 'end'] = f_split[3]
            file_df.loc[f_num, 'file_name'] = f_name+'.csv'
    file_df.reset_index(inplace=True,drop = True)
    return file_df

```

```python
all_df = get_weather_filelist()
all_df['file_name'] = './weather/' + all_df['file_name']
```

```python
for year in range(2010,2021):
    year_idx = str(year) + '01'
    year_df = all_df[all_df['start'] == year_idx]
    year_df.reset_index(inplace=True, drop = True)
    for dong in sorted(year_df['h_dong'].unique()):
        dong_year_df = year_df[year_df['h_dong'] == dong]
        if len(dong_year_df) != 5:
            print(f'{year} {dong}   {len(dong_year_df)}')
    print()
```
