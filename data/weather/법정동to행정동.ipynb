```python
import pandas as pd
import json
dong_nm_df = pd.read_excel('행정동_법정동_20221223.xlsx')
```

```python
cd_df = dong_nm_df[dong_nm_df['시군구명'] == '춘천시'][['법정동코드','읍면동명']]
```

```python
cd_dict = cd_df.set_index(['법정동코드']).to_dict()['읍면동명']
```

```python
dong_json = json.dumps(cd_dict)
```

```python
with open('dong_conv.json' , 'w') as json_file:
    json.dump(dong_json, json_file)
```

```python
load_dict = json.loads(dong_json)
```
