```python
import os
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
import pandas as pd
pd.reset_option('all')
pd.set_option('mode.chained_assignment',  None)
import numpy as np
from scipy.stats import mode
```

```python
def get_weather_filelist(path = './weather'):
    fs = sorted(os.listdir(path))
    file_df = pd.DataFrame(columns = ['h_dong', 'file_list' , 'start' , 'end', 'file_name'])

    for f_num , f_name in enumerate(fs):
        if 'ipy' in f_name:
            pass
        else:
            f_name = f_name.split('.')[0]
            f_split = f_name.split('_')
            file_df.loc[f_num, 'h_dong'] = f_split[0]
            file_df.loc[f_num, 'file_list'] = f_split[1]
            file_df.loc[f_num, 'start'] = f_split[2]
            file_df.loc[f_num, 'end'] = f_split[3]
            file_df.loc[f_num, 'file_name'] = f_name+'.csv'
    file_df.reset_index(inplace=True,drop = True)
    return file_df

def year_date_idx(year):
    return pd.date_range(f'{year}/01/01' ,f'{year+1}/01/01' , freq= '1h' )[:-1]


def formatting_weather(f_name , fn):
    dong_df = pd.DataFrame()
    dong_df.index = year_date_idx(year)

    df = pd.read_csv(f_name)
    init_df = pd.DataFrame(columns = df.columns)
    init_df.loc[0,df.columns[0]] = f' Start : {year}0101 '
    df = pd.concat([init_df, df])
    df.rename(columns = {df.columns[0] : 'day' ,df.columns[-1] : 'value' } , inplace=True)
    df.reset_index(inplace=True, drop = True)

    for idx in df.index:
        if df['day'].iloc[idx].find(' Start') == 0:
            month = int(df['day'].iloc[idx].split(':')[-1][5:7])
        df.loc[idx , 'month'] = int(month)
    tmp = df.groupby(['day', 'month' ,'hour'], as_index=False)
    group_df = tmp.agg(fn)
    group_df = group_df[['day' ,'month' , 'hour' , 'value']]
    if fn == mode:
        group_df['value'] = group_df['value'].apply(lambda x: x[0][0])
    
    group_df['month'] = group_df['month'].astype(int)
    group_df['day'] = group_df['day'].str.lstrip().astype(int)
    group_df['hour'] = group_df['hour'] /100
    group_df['hour'] = group_df['hour'].astype(int)
    
    for idx in group_df.index:
        month = group_df['month'].loc[idx]
        day = group_df['day'].loc[idx]
        hour = group_df['hour'].loc[idx]    
        if hour -1 == 1:
            dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour-2}:00:00') , 'value'] = group_df.iloc[idx]['value']

        dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour-1}:00:00') , 'value'] = group_df.iloc[idx]['value']
        dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour}:00:00') , 'value'] = group_df.iloc[idx]['value']

        if hour+1 < 24:
            dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour+1}:00:00') , 'value'] = group_df.iloc[idx]['value']
    return dong_df
```

```python
all_df = get_weather_filelist()
all_df
```

```python
def formatting_weather(f_name , year ,fn):
    dong_df = pd.DataFrame()
    dong_df.index = year_date_idx(year)
    
    df = pd.read_csv(f_name)
    init_df = pd.DataFrame(columns = df.columns)
    init_df.loc[0,df.columns[0]] = f' Start : {year}0101 '
    df = pd.concat([init_df, df])
    df.rename(columns = {df.columns[0] : 'day' ,df.columns[-1] : 'value' } , inplace=True)
    df.reset_index(inplace=True, drop = True)
    df.drop(['forecast'], axis=1 , inplace=True)
    for idx in df.index:
        if df['day'].iloc[idx].find(' Start') == 0:
            month = int(df['day'].iloc[idx].split(':')[-1][5:7])
        df.loc[idx , 'month'] = int(month)
    
    tmp = df.groupby(['day', 'month' ,'hour'], as_index=False)
    group_df = tmp.agg(fn)
    group_df = group_df[['day' ,'month' , 'hour' , 'value']]

    if fn == mode:
        group_df['value'] = group_df['value'].apply(lambda x: x[0][0])

    #if len(group_df) == 2922:
    #    group_df = group_df[1:-1]
    
    group_df['month'] = group_df['month'].astype(int)
    group_df['day'] = group_df['day'].str.lstrip().astype(int)
    group_df['hour'] = group_df['hour'].astype(float)
    group_df['hour'] = group_df['hour'] /100
    group_df['hour'] = group_df['hour'].astype(int)
    
    for idx in group_df.index:
        
        month = group_df['month'].loc[idx]
        day = group_df['day'].loc[idx]
        hour = group_df['hour'].loc[idx]
        value = group_df.iloc[idx]['value']
                
        if hour == 2:
            dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour-2}:00:00') , 'value'] = group_df.iloc[idx]['value']    
            #print(f"{month} {day} {hour-2} | {value}")
        if hour == 23:
            dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour-1}:00:00') , 'value'] = group_df.iloc[idx]['value']
            dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour}:00:00') , 'value'] = group_df.iloc[idx]['value']
            #print(f"{month} {day} {hour-1} | {value}")
            #print(f"{month} {day} {hour} | {value} ****")
            continue
        
        
        dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour-1}:00:00') , 'value'] = group_df.iloc[idx]['value']
        dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour}:00:00') , 'value'] = group_df.iloc[idx]['value']
        dong_df.loc[pd.to_datetime(f'{year}/{month}/{day} {hour+1}:00:00') , 'value'] = group_df.iloc[idx]['value']
        #print(f"{month} {day} {hour-1} | {value}")
        #print(f"{month} {day} {hour} | {value} ****")
        #print(f"{month} {day} {hour+1} | {value}")

        
    return dong_df


def year_weather_df(year , dong):
    year_df = all_df[all_df['start'] == f'{year}01']
    year_df['file_name'] = './weather/' + year_df['file_name']
    year_df.reset_index(inplace=True , drop=True)
    dong_year_df = year_df[year_df['h_dong']== dong]
    dong_weather = pd.DataFrame(columns= ['windspd','humid', 'temp', 'precip_form', 'precip'])
    
    temp_f_name = dong_year_df[dong_year_df['file_list'].isin(['3시간기온'])]['file_name'].iloc[0]
    precip_f_name = dong_year_df[dong_year_df['file_list'].isin(['12시간강수량' ,'6시간강수량' ])]['file_name'].iloc[0]
    precip_form_f_name = dong_year_df[dong_year_df['file_list'].isin(['강수형태'])]['file_name'].iloc[0]
    humid_f_name = dong_year_df[dong_year_df['file_list'].isin(['습도'])]['file_name'].iloc[0]    
    windspd_f_name = dong_year_df[dong_year_df['file_list'].isin(['풍속'])]['file_name'].iloc[0]
    
    #print(temp_f_name)
    #print(precip_f_name)
    #print(precip_form_f_name)
    #print(humid_f_name)
    #print(windspd_f_name)
    
    # 인자로 들어가는 arg 함수 수정 
    
    temp = formatting_weather(temp_f_name, year, np.mean)['value']
    precip = formatting_weather(precip_f_name, year, np.mean)['value']
    precip_form = formatting_weather(precip_form_f_name, year, mode)['value']
    humid = formatting_weather(humid_f_name, year, np.mean)['value']
    windspd = formatting_weather(windspd_f_name, year, mode)['value']

    
    dong_weather['temp'] = temp
    dong_weather['precip'] = precip
    dong_weather['precip_form'] = precip_form
    dong_weather['humid'] = humid
    dong_weather['windspd'] = windspd
    if sum(dong_weather.isnull().sum()) > 0:
        print(dong_weather.isnull().sum())
    
    dong_weather.rename_axis('REG_DTIME').reset_index(inplace=True)
    dong_weather = dong_weather.rename_axis('REG_DTIME').reset_index()
    return dong_weather

def year_df(year, dong):
    year_df = pd.DataFrame()
    tmp = year_weather_df(year, dong)
    tmp['h_dong'] = dong
    year_df = pd.concat([year_df , tmp])
    return year_df

def total_weather_df():
    years = range(2010, 2021)
    dongs = all_df['h_dong'].unique()
    t_df = pd.DataFrame()
    for year in years:
        for dong in dongs:
            print(f'{year} {dong}is processing')
            tmp_df = year_df(year, dong)
            t_df = pd.concat([t_df, tmp_df])
    return t_df
```
