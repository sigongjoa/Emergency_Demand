```python
# 라이브러리 , 데이터 load
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

plt.rcParams['font.family'] = 'NanumGothic'

data = pd.read_csv('total_data.csv')

data['REG_DTIME'] = pd.to_datetime(data['REG_DTIME'])
#target_dict = {0 : 0.1 , 1:0.9 , 2:1.9 , 3:2.9 , 4:3.9 , 5:4.9 , 6:5.9}
#data['count'] = data['count'].map(target_dict)

# sample data
data
```

```python
# functions
def dis_df(df):
    '''
    training dataset에서 time * count * dong으로 바꿔주는 함수 
    '''
    dt = pd.date_range(start, end ,freq='h')
    df = df[df['REG_DTIME'].isin(dt)]
    dongs = df['h_dong'].unique()
    ret_df = pd.DataFrame()
    for dong in dongs:
        ret_df[dong] = df[df['h_dong'] == dong]['count'].values
    ret_df.index = dt
    return ret_df

def moving_average(df, unit):
    '''
    training datset에 ewma를 적용하는 함수 
    '''
    ret_df = pd.DataFrame()
    max_value = df['count'].max()
    # forward df
    for dong in df['h_dong'].unique():
        dong_df = df[df['h_dong'] == dong]
        inv_dong_df = df[df['h_dong'] == dong][::-1]
        dong_df['count'] = dong_df['count'].ewm(span=unit , adjust=False).mean()
        for_ewma = dong_df['count'].ewm(span=unit , adjust=False).mean()
        back_ewma = inv_dong_df['count'].ewm(span=unit , adjust=False).mean()

        ewma = (for_ewma + back_ewma)/2 
        df['count'][dong_df.index] = ewma 
    df['count'] = df['count']  /  df['count'].max() * (max_value + 1.3)
    
    #print(max_value , df['count'].max())
    return df


mv_data = moving_average(data,3)
mv_data
data['count'].value_counts()
data
```

```python
start = '2020/1/1/00'
end = '2020/12/31/23'

data = pd.read_csv('total_data.csv')
data['REG_DTIME'] = pd.to_datetime(data['REG_DTIME'])

df1 = dis_df(data)

mv_data = data.copy(deep=True)
mv_data = moving_average(mv_data,3) #  window size 6
df2 = dis_df(mv_data)
data
```

```python
start = '2021/1/1/00'
end = '2021/12/31/23'


def plot_dong(dong):
    '''
    하나의 행정구역만 선택해서 위 start와 end를 이용해서 원하는 기간에 대한 출동 빈도수를 확인
    '''
    plt.figure(figsize=(15,4))
    plt.title(f'{start} ~ {end}' , size=20)

    dt = pd.date_range(start , end , freq = 'h')
    
    original_df = dis_df(data).loc[dt]
    emwa_df = dis_df(mv_data).loc[dt]
    
    org_dong_df = dis_df(data)[dong]
    ewma_dong_df = dis_df(mv_data)[dong]
    plt.plot(org_dong_df.index ,org_dong_df  , label='원본 데이터' , marker='o')
    plt.plot(ewma_dong_df.index ,ewma_dong_df, label = 'ewma(6h) 적용', marker='o')
    #plt.xticks(org_dong_df.index , rotation=45)
    plt.legend()

plot_dong('강남동')
# '동면'과 같이 2글자로된 동명은 '동  면'과 같이 중간에 띄어쓰기 2개를 추가
```

```python
# step function
def step(x, interval):
    for idx in range(len(interval)-1):
        if interval[idx] < x < interval[idx+1]:
            #print(interval[idx])
            return interval[idx]
    return interval[idx]    
```

```python
interval = np.linspace(0, 6, 40)
interval
step(8,interval)
```

```python
mv_data['count'] = mv_data['count'].apply(lambda x : step(x , interval))
```
