```python
# 동 별 f1 score
def f_name_split1(fn):
    return fn.split('/')[2][:-1] #, fn[3].split('.')[0] 
def f_name_split2(fn):
    return fn.split('/')[3].split('.')[0]

def confusion_matrix_plot(training, tft ,ewma_data, dong , title):
    fig = plt.figure(figsize=(25, 7))
    gs = gridspec.GridSpec(nrows=2, # row 몇 개 
                           ncols=2, # col 몇 개 
                           height_ratios=[1,1], 
                           width_ratios=[5,1]
                          )

    gs.update(wspace=0.025, hspace=0.2)
    ax0 = plt.subplot(gs[: , 0])
    ax1 = plt.subplot(gs[1,1])
    ax2 = plt.subplot(gs[0,1])

    ax1.set_xticks([])
    ax1.set_yticks([])
    ax1.axis('off')
    ax2.set_xticks([])
    ax2.set_yticks([])
    ax2.axis('off')

    ax0.set_title(f'{dong}_plot' , fontsize = 20)
    ax2.set_title('confusion_matrix', fontsize = 20)
    ax1.set_title('recall & precision , MAE', fontsize = 20)

    ewma_dong_df = ewma_data[ewma_data['h_dong']==dong]
    ewma_df_index = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['REG_DTIME']
    ewma_org_count = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['count']

    predition5= np.array([])
    for i in range(1,59):
        val_data = get_val_dataloader(training ,ewma_data, 8760+24*i)
        pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
        dong_idx = idx_df[idx_df['h_dong'] == dong].index[0]
        predition5 = np.concatenate([predition5,pred['prediction'][dong_idx, : , 5]])
    
    ax0.plot(ewma_df_index,predition5 , label = 'preidction' , alpha=0.4)

    org_df = data_processing('../../data/data.csv')
    dong_org_df = org_df[org_df['h_dong'] == dong]
    dong_org_count = dong_org_df[dong_org_df['time_idx'] > 8760-1]['count']

    dong_org_count = dong_org_count.to_numpy().round()
    dong_org_count = np.logical_not(dong_org_count < 1)
    predition5 = predition5.round()
    predition5 = np.logical_not(predition5 < 1)

    ax0.plot(ewma_df_index,dong_org_count , label = 'orginal target' , alpha=1)
    ax0.plot(ewma_df_index,predition5 ,label = '90% prediction round', alpha= 0.6, linestyle='--')
    ax0.legend()

    tn, fp, fn, tp = metrics.confusion_matrix(dong_org_count,predition5).ravel()

    c_mat = np.array([[tp , fp],[fn , tn]])
    ax1 = sns.heatmap(c_mat, annot=True, cbar = False, fmt='g',cmap='Blues')

    f1_score = metrics.f1_score(dong_org_count,predition5)
    recall = metrics.recall_score(dong_org_count,predition5)
    precision = metrics.precision_score(dong_org_count,predition5)
    MAE = abs(recall - precision)

    ax2.text(0.1, 3.1 , f'f1 score  {f1_score : 4f}' , fontsize=25)
    ax2.text(0.1, 3.4 , f'recall      {recall : 4f}' , fontsize=25)
    ax2.text(0.1, 3.7 , f'precision {precision : 4f}' , fontsize=25)
    ax2.text(0.1, 4.1 , f'MAE        {MAE : 4f}' , fontsize=25)
    
def dong_f1(training, tft ,ewma_data, dong):
    val_data = get_val_dataloader(training ,ewma_data, 8760+24)
    pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
    dong_idx = idx_df[idx_df['h_dong'] == dong].index[0]
    predictions = pred['prediction'][:, : , 5]
    
    for i in range(2,59):
        val_data = get_val_dataloader(training ,ewma_data, 8760+24*i)
        pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
        predictions = torch.cat([predictions ,pred['prediction'][:, : , 5]], dim=1) 
        dong_pred = predictions[dong_idx, :]

        
    ewma_dong_df = ewma_data[ewma_data['h_dong']==dong]
    ewma_df_index = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['REG_DTIME']
    ewma_org_count = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['count']
    
    org_df = data_processing('../../data/data.csv')
    dong_org_df = org_df[org_df['h_dong'] == dong]
    dong_org_count = dong_org_df[dong_org_df['time_idx'] > 8760-1]['count']
    
    dong_org_count = dong_org_count.to_numpy().round()
    dong_org_count = np.logical_not(dong_org_count < 1)
    dong_pred = dong_pred.round()
    dong_pred = np.logical_not(dong_pred < 1)
    return metrics.f1_score(dong_org_count,dong_pred)

def dong_f1_all(training , tft ,data):
    val_data = get_val_dataloader(training ,data, 8760+24)
    pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
    predictions = pred['prediction'][:, : , 5]

    for i in range(2,59):
        val_data = get_val_dataloader(training ,data, 8760+24*i)
        pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
        predictions = torch.cat([predictions ,pred['prediction'][:, : , 5]], dim=1) 
    predictions = predictions.reshape(-1,1)

    org_df = data_processing('../../data/data.csv')
    org_df = org_df[org_df['time_idx'] > 8760-1]
    org_list = []
    for dong in dongs:
        org_list.extend(org_df[org_df['h_dong'] == dong]['count'].to_list())

    org_list = np.array(org_list).round()
    org_list = np.logical_not(org_list < 1)
    predictions = predictions.round()
    predictions = np.logical_not(predictions < 1)

    return metrics.f1_score(org_list,predictions)

def get_f1_df(dong):
    data_sheet['f1_score'] = -1
    data_sheet.to_csv(f'{dong}_f1.csv', index=False)
        
    
def get_f1_score(dong):
    df= pd.read_csv(f'{dong}_f1.csv')
    for idx in df.index:
        if df['f1_score'].loc[idx] < 0:
            f_name = '../2022_11_02_TEST_ewma/alpha1/04.csv'
            model_ckpt = df['model_ckpt'].loc[idx]
            data = data_processing(f_name)
            train_data = data[data['time_idx']  < 8760]
            training = get_training(train_data,24, 24*7)
            tft = TemporalFusionTransformer.load_from_checkpoint(model_ckpt)
            if dong == 'all':
                df.loc[idx , 'f1_score'] = dong_f1_all(training, tft, data)
            else:
                df.loc[idx , 'f1_score'] = dong_f1(training, tft, data, dong)
            
            df.to_csv(f'{dong}_f1.csv',index=False)
            print(f'create {dong} f1 score with {idx}/{len(df)}')
            
def get_f1_result(idx):
    f_name = '../2022_11_02_TEST_ewma/alpha1/04.csv'
    model_ckpt = data_sheet['model_ckpt'].loc[idx]
    data = data_processing(f_name)
    train_data = data[data['time_idx']  < 8760]
    training = get_training(train_data,24, 24*7)
    tft = TemporalFusionTransformer.load_from_checkpoint(model_ckpt)
    for dong in dongs:
        print(dong)
        confusion_matrix_plot(training, tft, data, dong , 'BEST')

```
