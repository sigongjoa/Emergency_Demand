```python
# load module
import os
#os.environ['CUDA_LAUNCH_BLOCKING'] = "1"
os.environ["CUDA_VISIBLE_DEVICES"] = "0"

import copy
import warnings
import torch
import optuna
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pathlib import Path
from tqdm import tqdm

import pytorch_lightning as pl
import seaborn as sns

from my_funs import *
from sklearn import metrics
from matplotlib import gridspec
from pytorch_lightning.callbacks import EarlyStopping, LearningRateMonitor
from pytorch_lightning.loggers import TensorBoardLogger


from pytorch_forecasting import Baseline, TemporalFusionTransformer, TimeSeriesDataSet , EncoderNormalizer , GroupNormalizer
from pytorch_forecasting.metrics import SMAPE, PoissonLoss, QuantileLoss
#from pytorch_forecasting.models.temporal_fusion_transformer.tuning import optimize_hyperparameters

#import tensorflow as tf 
import tensorboard as tb 
#tf.io.gfile = tb.compat.tensorflow_stub.io.gfile

import matplotlib
matplotlib.rcParams['axes.unicode_minus'] = False
warnings.filterwarnings("ignore")  # avoid printing out absolute paths
plt.rcParams['font.family'] = 'NanumGothic'

dongs = ['강남동', '교  동', '근화동', '남  면', '남산면', '동  면', '동내면', '동산면', '북산면',
       '사북면', '서  면', '석사동', '소양동', '신동면', '신북읍', '신사우동', '약사명동', '조운동',
       '퇴계동', '효자1동', '후평1동']
```

```python
data_sheet = pd.read_csv('../2022_11_02_TEST_ewma/result_ewma_lr_optim.csv')
data_sheet['ewma'] = '../2022_11_02_TEST_ewma/' + data_sheet['ewma'] 
data_sheet['model_ckpt'] = '../2022_11_02_TEST_ewma/' + data_sheet['model_ckpt'] 
data_sheet
```

```python
# 동 별 f1 score
def f_name_split1(fn):
    return fn.split('/')[2][:-1] #, fn[3].split('.')[0] 
def f_name_split2(fn):
    return fn.split('/')[3].split('.')[0]

def confusion_matrix_plot(training, tft ,ewma_data, dong , title):
    fig = plt.figure(figsize=(25, 5)) 
    gs = gridspec.GridSpec(nrows=1, # row 몇 개 
                           ncols=2, # col 몇 개 
                           height_ratios=[1], 
                           width_ratios=[5,1]
                          )
    gs.update(wspace=0.025, hspace=0.05)
    ax0 = plt.subplot(gs[0])
    ax1 = plt.subplot(gs[1])
    
    val_data = get_val_dataloader(training ,ewma_data, 8760+24)
    pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
    dong_idx = idx_df[idx_df['h_dong'] == dong].index[0]
    
    predictions = pred['prediction'][:, : , 5]
    
    for i in range(2,59):
        val_data = get_val_dataloader(training ,ewma_data, 8760+24*i)
        pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
        predictions = torch.cat([predictions ,pred['prediction'][:, : , 5]], dim=1) 
        dong_pred = predictions[dong_idx, :]

        
    ewma_dong_df = ewma_data[ewma_data['h_dong']==dong]
    ewma_df_index = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['REG_DTIME']
    ewma_org_count = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['count']
    ax0.plot(ewma_df_index,dong_pred , label = '90% prediction' , alpha=0.6 , color = 'violet')
    
    org_df = data_processing('../../data/data.csv')
    dong_org_df = org_df[org_df['h_dong'] == dong]
    dong_org_count = dong_org_df[dong_org_df['time_idx'] > 8760-1]['count']
    
    dong_org_count = dong_org_count.to_numpy().round()
    dong_org_count = np.logical_not(dong_org_count < 1)
    dong_pred = dong_pred.round()
    dong_pred = np.logical_not(dong_pred < 1)
    
    ax0.plot(ewma_df_index,dong_org_count , label = 'orginal target' , alpha=1)
    ax0.plot(ewma_df_index,ewma_org_count , label = 'ewma target' , alpha=0.5)
    ax0.plot(ewma_df_index,dong_pred ,label = '90% prediction round', alpha= 0.6, linestyle='--')
    ax0.set_title(f'{dong}_{title}')
    ax0.legend()
    
    tn, fp, fn, tp = metrics.confusion_matrix(dong_org_count,dong_pred).ravel()

    c_mat = np.array([[tp , fp],[fn , tn]])
    ax1 = sns.heatmap(c_mat, annot=True, cbar = False, fmt='g',cmap='Blues')
    ax1.set_title('confusion_matrix')
    ax1.set_xticks([])
    ax1.set_yticks([])
    print(f'{dong} f1_score' , metrics.f1_score(dong_org_count,dong_pred))
    return metrics.f1_score(dong_org_count,dong_pred)

def dong_f1(training, tft ,ewma_data, dong):
    val_data = get_val_dataloader(training ,ewma_data, 8760+24)
    pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
    dong_idx = idx_df[idx_df['h_dong'] == dong].index[0]
    predictions = pred['prediction'][:, : , 5]
    
    for i in range(2,59):
        val_data = get_val_dataloader(training ,ewma_data, 8760+24*i)
        pred , x, idx_df = tft.predict(val_data, mode='raw', return_x = True , return_index = True)
        predictions = torch.cat([predictions ,pred['prediction'][:, : , 5]], dim=1) 
        dong_pred = predictions[dong_idx, :]

        
    ewma_dong_df = ewma_data[ewma_data['h_dong']==dong]
    ewma_df_index = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['REG_DTIME']
    ewma_org_count = ewma_dong_df[ewma_dong_df['time_idx'] > 8760-1]['count']
    
    org_df = data_processing('../../data/data.csv')
    dong_org_df = org_df[org_df['h_dong'] == dong]
    dong_org_count = dong_org_df[dong_org_df['time_idx'] > 8760-1]['count']
    
    dong_org_count = dong_org_count.to_numpy().round()
    dong_org_count = np.logical_not(dong_org_count < 1)
    dong_pred = dong_pred.round()
    dong_pred = np.logical_not(dong_pred < 1)
    return metrics.f1_score(dong_org_count,dong_pred)

def get_f1_df(dong):
    data_sheet['f1_score'] = -1
    data_sheet.to_csv(f'{dong}_f1.csv', index=False)
    
def get_f1_score(dong):
    df= pd.read_csv(f'{dong}_f1.csv')
    for idx in df.index:
        if df['f1_score'].loc[idx] < 0:
            f_name = df['ewma'].loc[idx]
            model_ckpt = df['model_ckpt'].loc[idx]
            data = data_processing(f_name)
            train_data = data[data['time_idx']  < 8760]
            training = get_training(train_data,24, 24*7)
            tft = TemporalFusionTransformer.load_from_checkpoint(model_ckpt)
            df.loc[idx , 'f1_score'] = dong_f1(training, tft, data, dong)
            df.to_csv(f'{dong}_f1.csv',index=False)
            print(f'create {dong} f1 score with {idx}/125')
            
def get_f1_result(dong, n):
    df= pd.read_csv(f'{dong}_f1.csv')
    df['factor'] = df['ewma'].apply(lambda x : f_name_split2(x))
    df['ewma'] = df['ewma'].apply(lambda x : f_name_split1(x))
    df = df[['ewma' ,'factor', 'lr' ,'optim' , 'f1_score']]
    df = df.sort_values(by='f1_score', ascending=False)
    
    for idx in df.index[:n]:
        f_name = data_sheet['ewma'].loc[idx]
        model_ckpt = data_sheet['model_ckpt'].loc[idx]
        #print(f_name, model_ckpt)
        data = data_processing(f_name)
        train_data = data[data['time_idx']  < 8760]
        training = get_training(train_data,24, 24*7)
        tft = TemporalFusionTransformer.load_from_checkpoint(model_ckpt)
        confusion_matrix_plot(training, tft, data, '강남동' , 'test')
    return df
```
