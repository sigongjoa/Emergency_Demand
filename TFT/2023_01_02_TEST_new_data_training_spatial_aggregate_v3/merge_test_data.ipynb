```python
nei_data[~nei_data['REG_DTIME'].isin(dt)].to_csv('nei_test.csv',index=False)
nei_data[~nei_data['REG_DTIME'].isin(dt)]
```

```python
test_nei_data = nei_data[~nei_data['REG_DTIME'].isin(dt)]
for x , y in enumerate(test_nei_data['REG_DTIME'].unique()):
    test_nei_data_index = test_nei_data[test_nei_data['REG_DTIME'] == y].index
    print(test_nei_data_index , x)
    test_nei_data.loc[test_nei_data_index , 'time_idx'] = x
```

```python
for idx in range(52):
    if idx < 10:
        test_nei_data[test_nei_data['time_idx'].isin(range(idx*24 , 24*(idx+7)))].to_csv(f'test_data/0{idx}_data.csv',index=False)
    else:
        test_nei_data[test_nei_data['time_idx'].isin(range(idx*24 , 24*(idx+7)))].to_csv(f'test_data/{idx}_data.csv',index=False)
```
